### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv')
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId',inplace=True,drop=True)
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
# 1 is Male and 0 is Female
titanic['Sex_int'] = titanic.Sex.apply(lambda x:1 if x=='male' else 0)
temp = titanic.Sex_int.value_counts()
plt.pie(temp.values,labels=['Males','Females'],autopct='%.1f%%');
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.scatterplot(x='Age',y='Fare',hue=titanic.Sex.values.tolist(),data=titanic);
```

### Step 7. How many people survived?

```python
titanic.Survived.sum()
```

### Step 8. Create a histogram with the Fare payed

```python
sns.distplot(titanic.Fare,kde=False);
# plt.hist(titanic.Fare,range=[0,50])
```

### Correlation of Features with Response

```python
titanic[['Pclass','Age','SibSp','Parch','Fare','Sex_int']].corrwith(titanic.Survived).plot.bar(figsize=(20,20),
                                                                                               title='Correlation with Response',
                                                                                              colormap='Paired',fontsize=20,
                                                                                              grid=True);
```

##### As Female is 0 and Male is 1, and by plot we can see that survival rate is increasing as Sex_int is decreasing (they are negatively correlated) so Fermale are more likely to survive than Men.
##### Also people paying higher Fare (thus better facilities and higher class) are more likely to survive than people paying lower Fare (thus lower class)
