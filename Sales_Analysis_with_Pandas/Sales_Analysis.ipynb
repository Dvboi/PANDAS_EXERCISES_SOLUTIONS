##### Merging 12 months of sales into single data file

```python
import glob   # regex for file navigation
import os
path = r'.\Sales_Data'
all_files = glob.glob(os.path.join(path,'*.csv')) # get every file in Sales_Data folder with .csv extension
li = []
for filename in all_files:
    df = pd.read_csv(filename,parse_dates=['Order Date'])
    li.append(df)
```

```python
data = pd.concat(li,axis=0)
data.head()
```

```python
print("Shape of dataset:")
print(data.shape,end='\n\n')
print('-'*130)

print("Dtypes:")
print(data.dtypes,end='\n\n')
print('-'*130)

print("Nulls:")
print(data.isnull().sum(),end='\n\n')
print('-'*130)
```

```python
data[data.isna().any(axis=1)]
# WHOLE ROWS HAVE NULL SO WE CAN REMOVE IT
```

```python
data.dropna(inplace=True)
print(data.shape)
print('-'*130)
# also many columns contains just column names
print(data[data['Order ID']=='Order ID'])
print('-'*130)
# dropping those 
data = data[data['Order ID']!='Order ID']
#resetting indexes
data.reset_index(drop=True,inplace=True)
# convert data-types
convert_dict = {'Order ID': int,
 'Product': 'string',
 'Quantity Ordered': int,
 'Price Each': float,
 'Purchase Address': 'string'}
data = data.astype(convert_dict)
# convert to date-time separately
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
# exporting clean dataset
data.to_csv('Final_data.csv',index=False)
del data  # removing from memory
```

```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Final_data.csv',parse_dates = ['Order Date'])
```

```python
df.head()
```

Month with highest sales?


```python
df['sp'] = df['Price Each']*df['Quantity Ordered']
plt.figure()
df.groupby(df['Order Date'].dt.month)['sp'].sum().plot.bar()
plt.xlabel('Month No.');
# Month 12 aka December  (maybe cuz christmas and s)
```

City with Highest sales?


```python
import re
def clean(x):
    return x.split(',')[1] + ',' + re.sub(r'[0-9]','',x.split(',')[2])
temp = df.groupby(df['Purchase Address'].apply(lambda x:clean(x)))['sp'].sum()  

# plotting
plt.figure()
sns.barplot(x = temp.index,y=temp.values )
plt.xticks(rotation = 45,rotation_mode = "anchor",ha = "right")
plt.subplots_adjust(bottom=0.25)
```

```python
## San Francisco
```

BEST TIME FOR ADVERTISEMENTS

```python
# Assuming ORDER ID are valid
temp = df.groupby(df['Order Date'].dt.hour)['Order ID'].nunique()
hours = [i for i,_ in df.groupby(df['Order Date'].dt.hour)]

plt.figure()
plt.plot(temp)
plt.xticks(hours)
plt.grid()
plt.xlabel('Hour of DAY')
plt.ylabel('Count of Order')
```

Bimodal distribution showing different times of peak in order counts
* 0 a.m. - 7 a.m. --> Low orders 
* 12 p.m. & 7 p.m. -> Max orders
> Probably at time of lunch periods and after coming from offices

Which Products are most often sold together

```python
temp = df.groupby('Order ID')['Product'].apply(lambda x:','.join(x))
temp = pd.DataFrame(temp)
temp = temp[temp.Product.str.contains(',')]
top_10 = temp.groupby('Product')['Product'].count().sort_values(ascending=False).head(10)

#visualize

plt.figure()
top_10.plot(kind='barh')
plt.xticks([0,500,1000])
plt.subplots_adjust(left=0.70)
```

```python
## ALTERNATE WAY , more efficient too!
# for combinations of n-type
from itertools import combinations
from collections import Counter
count = Counter()
for ele in temp['Product']:
    row = ele.split(',')
    count.update(Counter(combinations(row,2)))
top_10_new = count.most_common(10)

# Convert to series

d = {}
for k,v in top_10_new:
    k = ','.join(k)
    d[k] = v
top_10_new = pd.Series(d)

#visualize
plt.figure()
top_10_new.plot(kind='barh')
plt.xticks([0,500,1000])
plt.subplots_adjust(left=0.70)

```
